```python
import matplotlib.pyplot as plt
x1_values = [1, 2, 3, 4, 5]
y1_values = [2, 4, 6, 8, 10]
x2_values = [3, 6, 9, 12, 15]
y2_values = [2, 4, 6, 8, 10]
plt.figure(figsize=(10, 6))
plt.plot(x1_values, y1_values, label='Quantity 1', color='blue', marker='*')
plt.plot(x2_values, y2_values, label='Quantity 2', color='red', marker='o')
plt.title('Plot of Two Quantities with X and Y Positions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.tight_layout()
plt.show()
```
