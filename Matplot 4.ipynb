```python
import matplotlib.pyplot as plt
x_values = [1, 2, 3, 4, 5]
y1_values = [1, 3, 5, 7, 9]
y2_values = [2, 4, 6, 8, 10]
plt.figure(figsize=(10, 6))
plt.plot(x_values, y1_values, label='Line 1', color='blue', linewidth=3)
plt.plot(x_values, y2_values, label='Line 2', color='red', linewidth=5)
plt.title('Lines with different color and width')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.tight_layout()
plt.show()
```
